# Variables and Types

## Your first data packet from Mars

Somewhere north of 225 million kilometres away, the Curiosity rover is trundling across Gale Crater on Mars. Every Martian day (called a **sol**), it bundles up temperature readings, wind speeds, atmospheric pressure, and instrument diagnostics, then beams them back to Earth via a relay satellite.

You have just joined NASA's Jet Propulsion Laboratory as a data intern, and your first job is straightforward: take one sol of raw telemetry and make sense of it using Python.

Before we can do anything clever with this data, we need to understand how Python stores and handles different kinds of information. That is what this notebook is about.

## Storing telemetry in variables

A **variable** is just a name that points to a piece of data. Think of it as a labelled box: the label is the variable name, and the thing inside is the value.

Let's store some basic telemetry from Sol 142.

````python
sol_number = 142
min_temp_celsius = -63
max_temp_celsius = -8
wind_speed_ms = 14.7
instrument_status = "nominal"
dust_storm_active = False
````

Nothing printed, but Python has quietly stored six pieces of information. We can inspect any of them by simply typing the variable name.

````python
sol_number
````

````python
instrument_status
````

A few naming rules worth knowing:

- Variable names can contain letters, digits, and underscores, but cannot start with a digit.
- Python convention is `snake_case` for variables: all lowercase, words separated by underscores.
- Names are case-sensitive. `Sol_Number` and `sol_number` are entirely different variables.

## The four fundamental types

Every value in Python has a **type**. The type determines what operations are allowed. You wouldn't try to divide a sentence by three, and Python won't let you either.

For our telemetry data, we need four types.

````python
# int: whole numbers, no decimal point
print(type(sol_number))       # Sol 142 is a whole number

# float: numbers with a decimal point
print(type(wind_speed_ms))    # 14.7 m/s needs decimal precision

# str: text, always in quotes
print(type(instrument_status))  # "nominal" is text

# bool: True or False, nothing else
print(type(dust_storm_active))  # either there's a storm or there isn't
````

Notice that `type()` is a **function** -- we pass it a value, and it tells us what kind of value it is. We will write our own functions in a later notebook, but for now it is enough to know that functions do things for us.

Here is a quick reference:

| Type | Example | Use case |
|------|---------|----------|
| `int` | `142` | Counting, indexing |
| `float` | `14.7` | Measurements, calculations |
| `str` | `"nominal"` | Labels, text data |
| `bool` | `True` | Yes/no flags, conditions |

## Arithmetic with numbers

Python handles the usual mathematical operations, with a couple of useful extras.

````python
# Temperature range for Sol 142
temp_range = max_temp_celsius - min_temp_celsius
print("Temperature range:", temp_range, "degrees")
````

````python
# Convert Celsius to Fahrenheit
min_temp_f = (min_temp_celsius * 9/5) + 32
print("Min temp in Fahrenheit:", min_temp_f)
````

````python
# Some handy operators
print("Division:", 7 / 2)       # Always gives a float
print("Integer division:", 7 // 2)  # Rounds down to whole number
print("Remainder:", 7 % 2)      # Modulo -- the leftover
print("Power:", 2 ** 10)        # Exponentiation
````

A gotcha: regular division (`/`) always returns a float, even when the result is a whole number. If you need an integer, use `//`.

## f-strings: building telemetry reports

Raw numbers are useful for calculations, but at some point we need to produce human-readable reports. Python's **f-strings** (formatted string literals) make this painless.

Put an `f` before the opening quote, and then anything inside curly braces `{}` gets evaluated as Python code.

````python
report = f"Sol {sol_number}: min {min_temp_celsius}C, max {max_temp_celsius}C, wind {wind_speed_ms} m/s"
print(report)
````

````python
# You can put expressions inside the braces, not just variable names
print(f"Temperature range: {max_temp_celsius - min_temp_celsius} degrees")
print(f"Min temp in Fahrenheit: {(min_temp_celsius * 9/5) + 32:.1f}F")
````

That `:.1f` is a **format specifier** -- it means "show one decimal place". Some useful ones:

| Specifier | Meaning | Example |
|-----------|---------|----------|
| `:.2f` | 2 decimal places | `3.14` |
| `:.0f` | No decimal places | `3` |
| `:,` | Thousands separator | `1,000,000` |
| `:.1%` | Percentage | `73.2%` |

````python
# A more polished telemetry summary
pressure_pa = 735.4
uv_index = 7

summary = f"""=== SOL {sol_number} TELEMETRY REPORT ===
Temperature: {min_temp_celsius}C to {max_temp_celsius}C (range: {max_temp_celsius - min_temp_celsius}C)
Wind speed:  {wind_speed_ms:.1f} m/s
Pressure:    {pressure_pa:,.1f} Pa
UV index:    {uv_index}
Status:      {instrument_status.upper()}
Dust storm:  {'YES' if dust_storm_active else 'NO'}"""

print(summary)
````

The triple-quoted string (`""".."""`) lets us write multi-line text without fuss. Very handy for reports.

## The type contract: why "-63" + 5 crashes

Here is the single most important idea in this notebook.

Telemetry data arrives from Mars as radio signals, which get decoded into text. That means temperatures, pressures, wind speeds -- they all start life as **strings**, not numbers. Watch what happens when we try to do maths with a string that looks like a number.

````python
# This is what the raw telemetry looks like: everything is text
raw_temp = "-63"
print(type(raw_temp))  # It's a str, not an int
````

````python
# This will crash
try:
    result = raw_temp + 5
except TypeError as e:
    print(f"Error: {e}")
````

Python is telling us: *you cannot concatenate a string and an integer*. The `+` operator means different things for different types. For numbers, it means addition. For strings, it means gluing them together. Python refuses to guess which one you meant.

This is not a flaw. This is a feature. Types are **contracts** -- they tell Python (and you) what operations make sense for a given piece of data. A string that says `"-63"` is just five characters, not a number you can do arithmetic with.

The fix is **type conversion**.

````python
# Convert the string to an integer first
actual_temp = int(raw_temp)
print(type(actual_temp))
print(actual_temp + 5)  # Now this works: -63 + 5 = -58
````

````python
# The conversion functions
print(int("42"))        # str -> int
print(float("14.7"))    # str -> float
print(str(142))         # int -> str
print(bool(1))          # int -> bool (0 is False, everything else is True)
````

````python
# Processing a raw telemetry line
raw_line = "-63, -8, 14.7, 735.4, nominal"

# Split into parts (we'll learn more about .split() shortly)
parts = raw_line.split(", ")
print(parts)
````

````python
# Every item is a string -- we need to convert the numeric ones
min_temp = int(parts[0])
max_temp = int(parts[1])
wind = float(parts[2])
pressure = float(parts[3])
status = parts[4]

print(f"Min: {min_temp}C, Max: {max_temp}C, Wind: {wind} m/s")
````

This pattern -- receive text, split it, convert to the right types -- is something you will do constantly in data engineering. It is worth getting comfortable with.

## String methods: cleaning up messy telemetry

Real telemetry data is never perfectly clean. There are extra spaces, inconsistent capitalisation, stray characters. Python strings have a rich set of **methods** (functions that belong to the string) for dealing with this.

A method is called with dot notation: `my_string.method_name()`.

````python
# Some messy telemetry strings
raw_status = "  Nominal  "
raw_reading = "SOL:142|TEMP:-63|WIND:14.7"
raw_note = "dust storm WARNING issued at 14:32 UTC"
````

````python
# .strip() removes leading and trailing whitespace
clean_status = raw_status.strip()
print(f"'{raw_status}' -> '{clean_status}'")
````

````python
# .lower() and .upper() for consistent casing
print(clean_status.lower())   # "nominal"
print(clean_status.upper())   # "NOMINAL"
````

````python
# .split() breaks a string into a list of parts
fields = raw_reading.split("|")
print(fields)
````

````python
# We can split further to extract the values
for field in fields:
    key, value = field.split(":")
    print(f"  {key} = {value}")
````

````python
# .replace() swaps one substring for another
cleaned = raw_reading.replace("|", ", ")
print(cleaned)
````

````python
# .startswith() and .endswith() are useful for checking patterns
print(raw_note.startswith("dust"))  # True
print(raw_note.endswith("UTC"))     # True
````

````python
# .find() returns the position of a substring (-1 if not found)
pos = raw_note.find("WARNING")
print(f"'WARNING' found at position {pos}")
print(f"Not found example: {raw_note.find('ERROR')}")
````

An important detail: string methods **return a new string**. They never modify the original. Strings in Python are **immutable** -- once created, they cannot be changed. Every operation gives you a fresh string.

````python
# The original is unchanged
print(f"Original: '{raw_status}'")
raw_status.strip()  # This creates a new string but we don't store it
print(f"Still: '{raw_status}'")  # Unchanged!
````

## Chaining methods

Because each method returns a new string, we can chain them together. This is a common pattern for cleaning data in one readable line.

````python
messy = "  WARNING: High Wind Speed Detected  "
clean = messy.strip().lower().replace("warning: ", "")
print(clean)  # "high wind speed detected"
````

## Booleans and comparisons

We briefly met `bool` earlier. Booleans become essential when we start making decisions about data -- is this reading anomalous? Is the instrument online?

````python
# Comparison operators produce booleans
print(min_temp_celsius < -50)    # Is it extremely cold? True
print(wind_speed_ms > 25)        # Dangerous wind? False
print(instrument_status == "nominal")  # All systems go? True
print(sol_number != 100)         # Not sol 100? True
````

````python
# Combine with and, or, not
is_safe = instrument_status == "nominal" and not dust_storm_active
print(f"Safe to operate: {is_safe}")

needs_attention = wind_speed_ms > 20 or instrument_status != "nominal"
print(f"Needs attention: {needs_attention}")
````

## Putting it together

Let's combine everything we have learned to process a realistic telemetry packet.

````python
# A raw telemetry packet as it arrives from Mars (all strings)
raw_packet = """SOL: 142
DATE: 2024-06-15
MIN_TEMP: -63
MAX_TEMP: -8
PRESSURE: 735.4
WIND: 14.7
UV: 7
STATUS: nominal"""

print(raw_packet)
````

````python
# Parse each line into a key-value pair
for line in raw_packet.strip().split("\n"):
    key, value = line.split(": ", 1)  # Split on first ": " only
    print(f"  Key: {key:12s}  Value: {value}")
````

````python
# Full parsing with type conversion
data = {}
for line in raw_packet.strip().split("\n"):
    key, value = line.split(": ", 1)
    data[key] = value.strip()

# Convert numeric fields to proper types
sol = int(data["SOL"])
earth_date = data["DATE"]
min_t = int(data["MIN_TEMP"])
max_t = int(data["MAX_TEMP"])
pressure = float(data["PRESSURE"])
wind = float(data["WIND"])
uv = int(data["UV"])
status = data["STATUS"]

print(f"Parsed Sol {sol} ({earth_date}): {min_t}C to {max_t}C, {wind} m/s wind")
````

## Summary

In this notebook we have covered:

- **Variables** store data with meaningful names
- **Types** (`int`, `float`, `str`, `bool`) determine what operations are valid
- **f-strings** let us build formatted text with embedded expressions
- **Type conversion** (`int()`, `float()`, `str()`) is essential when data arrives as text
- **String methods** (`.strip()`, `.split()`, `.upper()`, `.replace()`) are our first tools for cleaning data

The big idea: types are contracts. When telemetry arrives from Mars as strings, we must explicitly convert to the right type before doing any calculations. Python will not guess for us, and that is a good thing.
