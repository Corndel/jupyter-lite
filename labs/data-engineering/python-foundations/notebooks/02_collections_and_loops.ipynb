# Collections and Loops

## A week of Mars weather

A week of sol data has arrived. Seven days of temperatures, wind speeds, and instrument readings from Curiosity's weather station. One variable per measurement was fine for a single sol, but we now have seven of everything, and creating `temp_sol_1`, `temp_sol_2`, `temp_sol_3`... would be absurd.

We need **collections** -- data structures that hold multiple values -- and **loops** to process them efficiently. This notebook introduces the tools that turn individual measurements into workable datasets.

## Lists: ordered sequences

A **list** is an ordered sequence of values, created with square brackets. It is the workhorse of Python data structures.

```python
# Seven sols of minimum temperatures (Celsius)
min_temps = [-63, -71, -58, -67, -74, -61, -69]
print(min_temps)
print(f"We have {len(min_temps)} readings")
```

```python
# Indexing: access individual items (0-based!)
print(f"First sol: {min_temps[0]}C")
print(f"Last sol:  {min_temps[-1]}C")  # Negative index counts from the end
print(f"Third sol: {min_temps[2]}C")
```

Zero-based indexing trips everyone up at first. The first element is at position 0, not 1. This is a convention Python inherited from C, and there are sensible mathematical reasons for it, but honestly you just get used to it.

```python
# Slicing: grab a range of elements
first_three = min_temps[0:3]   # Items at index 0, 1, 2 (not 3!)
last_three = min_temps[-3:]    # Last three items
middle = min_temps[2:5]        # Items at index 2, 3, 4

print(f"First three: {first_three}")
print(f"Last three:  {last_three}")
print(f"Middle:      {middle}")
```

The slicing rule is `[start:stop]` where `start` is included and `stop` is excluded. This sounds odd, but it means `len(my_list[a:b])` is always `b - a`, which turns out to be very convenient.

```python
# Lists are mutable -- we can change them
wind_speeds = [14.7, 8.3, 22.1, 5.6, 18.9, 11.2, 27.4]

# Oops, the sol 4 reading was wrong. Fix it:
wind_speeds[3] = 6.1
print(wind_speeds)
```

```python
# Adding to a list
wind_speeds.append(15.3)  # New sol 8 reading arrives
print(f"Now {len(wind_speeds)} readings: {wind_speeds}")
```

```python
# Useful list operations
print(f"Min temp this week:  {min(min_temps)}C")
print(f"Max temp this week:  {max(min_temps)}C")
print(f"Total wind readings: {sum(wind_speeds):.1f} m/s")
print(f"Average wind speed:  {sum(wind_speeds) / len(wind_speeds):.1f} m/s")
```

```python
# Checking membership
print(-63 in min_temps)    # True -- this temperature was recorded
print(-100 in min_temps)   # False -- this one was not
```

## Dictionaries: labelled data

Lists are great when order matters and items are similar (a sequence of temperatures, say). But a single sol has several different readings -- temperature, wind, pressure, status. We need a way to label each value.

A **dictionary** (dict) stores key-value pairs. Think of it as a lookup table.

```python
# One sol's complete data
sol_142 = {
    "sol": 142,
    "earth_date": "2024-06-15",
    "min_temp_c": -63,
    "max_temp_c": -8,
    "pressure_pa": 735.4,
    "wind_speed_ms": 14.7,
    "uv_index": 7,
    "instrument_status": "nominal"
}

print(sol_142)
```

```python
# Access values by key
print(f"Sol {sol_142['sol']}: {sol_142['min_temp_c']}C to {sol_142['max_temp_c']}C")
print(f"Status: {sol_142['instrument_status']}")
```

```python
# .get() is safer -- returns a default if the key doesn't exist
print(sol_142.get("dust_storm", "no data"))  # Key doesn't exist, so we get the default
print(sol_142.get("sol", "no data"))          # Key exists, so we get the value
```

```python
# Adding and updating entries
sol_142["notes"] = "Slight haze observed"  # New key
sol_142["uv_index"] = 8                     # Update existing key

print(sol_142["notes"])
print(sol_142["uv_index"])
```

```python
# Iterating over a dict
print("Keys:", list(sol_142.keys()))
print("Values:", list(sol_142.values()))

# .items() gives you both
for key, value in sol_142.items():
    print(f"  {key}: {value}")
```

## if/elif/else: making decisions

With real telemetry data, we need to flag anomalous readings. Is the temperature dangerously low? Is the wind speed too high? Is an instrument offline?

Conditional statements let us make decisions based on data.

```python
temp = -63

if temp < -70:
    print("ALERT: Extreme cold detected")
elif temp < -50:
    print("Normal Mars temperature range")
else:
    print("Unusually warm for Mars")
```

```python
# A more complete anomaly checker
def check_reading(sol_data):
    """Check a sol's data for anomalies and print warnings."""
    sol = sol_data["sol"]
    
    if sol_data["min_temp_c"] < -75:
        print(f"Sol {sol}: COLD ALERT -- min temp {sol_data['min_temp_c']}C")
    
    if sol_data["wind_speed_ms"] > 25:
        print(f"Sol {sol}: WIND ALERT -- {sol_data['wind_speed_ms']} m/s")
    
    if sol_data["instrument_status"] != "nominal":
        print(f"Sol {sol}: INSTRUMENT {sol_data['instrument_status'].upper()}")
    
    if sol_data["uv_index"] > 9:
        print(f"Sol {sol}: HIGH UV -- index {sol_data['uv_index']}")

check_reading(sol_142)
```

## for loops: processing sequences

The real power of collections comes when we combine them with loops. A `for` loop runs a block of code once for each item in a sequence.

```python
# Process each temperature reading
for temp in min_temps:
    status = "ALERT" if temp < -70 else "OK"
    print(f"  {temp}C  [{status}]")
```

```python
# enumerate() gives us the index alongside the value
for i, temp in enumerate(min_temps):
    print(f"  Sol {i+1}: {temp}C")
```

```python
# range() generates a sequence of numbers
for i in range(5):
    print(i, end=" ")  # 0 1 2 3 4

print()  # newline

for i in range(1, 8):  # 1 through 7
    print(f"Sol {i}", end="  ")
```

```python
# Building a list with a loop
temp_ranges = []
max_temps = [-8, -4, -15, -2, -18, -6, -12]

for i in range(len(min_temps)):
    temp_range = max_temps[i] - min_temps[i]
    temp_ranges.append(temp_range)

print(f"Temperature ranges: {temp_ranges}")
print(f"Average range: {sum(temp_ranges) / len(temp_ranges):.1f} degrees")
```

## List comprehensions: the Pythonic shortcut

That last pattern -- create empty list, loop, append -- is so common that Python has a compact syntax for it: the **list comprehension**.

```python
# The loop version:
temp_ranges_loop = []
for i in range(len(min_temps)):
    temp_ranges_loop.append(max_temps[i] - min_temps[i])

# The list comprehension version:
temp_ranges_comp = [max_temps[i] - min_temps[i] for i in range(len(min_temps))]

print(temp_ranges_loop)
print(temp_ranges_comp)
print(f"Same result: {temp_ranges_loop == temp_ranges_comp}")
```

```python
# Even cleaner with zip(), which pairs up elements from two lists
temp_ranges_zip = [mx - mn for mn, mx in zip(min_temps, max_temps)]
print(temp_ranges_zip)
```

```python
# Filtering with a condition
cold_days = [t for t in min_temps if t < -65]
print(f"Days below -65C: {cold_days}")

# Convert Celsius to Fahrenheit for all readings
min_temps_f = [(t * 9/5) + 32 for t in min_temps]
print(f"Fahrenheit: {[f'{t:.1f}' for t in min_temps_f]}")
```

List comprehensions are one of Python's best features. They are concise, readable (once you learn the pattern), and faster than the loop-and-append approach. Use them whenever you are transforming or filtering a list.

## Nested data: a week of sols

In practice, each sol is a dict with multiple readings, and a week of sols is a **list of dicts**. This nested structure is extremely common in data engineering -- it is essentially what a database table looks like before you load it into a proper tool.

```python
week_data = [
    {"sol": 142, "min_temp_c": -63, "max_temp_c": -8,  "wind_speed_ms": 14.7, "status": "nominal"},
    {"sol": 143, "min_temp_c": -71, "max_temp_c": -4,  "wind_speed_ms": 8.3,  "status": "nominal"},
    {"sol": 144, "min_temp_c": -58, "max_temp_c": -15, "wind_speed_ms": 22.1, "status": "warning"},
    {"sol": 145, "min_temp_c": -67, "max_temp_c": -2,  "wind_speed_ms": 6.1,  "status": "nominal"},
    {"sol": 146, "min_temp_c": -74, "max_temp_c": -18, "wind_speed_ms": 18.9, "status": "nominal"},
    {"sol": 147, "min_temp_c": -61, "max_temp_c": -6,  "wind_speed_ms": 27.4, "status": "offline"},
    {"sol": 148, "min_temp_c": -69, "max_temp_c": -12, "wind_speed_ms": 15.3, "status": "nominal"},
]
```

```python
# Access nested values
print(f"Sol {week_data[0]['sol']}: {week_data[0]['min_temp_c']}C")

# Process the whole week
for day in week_data:
    temp_range = day["max_temp_c"] - day["min_temp_c"]
    print(f"Sol {day['sol']}: {day['min_temp_c']}C to {day['max_temp_c']}C (range: {temp_range}C) [{day['status']}]")
```

```python
# Extract all min temps using a list comprehension
all_min_temps = [day["min_temp_c"] for day in week_data]
print(f"Min temps: {all_min_temps}")
print(f"Coldest sol: {min(all_min_temps)}C")
print(f"Average min: {sum(all_min_temps) / len(all_min_temps):.1f}C")
```

```python
# Find anomalous sols
anomalies = [
    day for day in week_data
    if day["min_temp_c"] < -70
    or day["wind_speed_ms"] > 25
    or day["status"] != "nominal"
]

print(f"Found {len(anomalies)} anomalous sols:")
for day in anomalies:
    print(f"  Sol {day['sol']}: temp={day['min_temp_c']}C, wind={day['wind_speed_ms']}m/s, status={day['status']}")
```

## Why dicts exist: the lookup speed revelation

So far, dicts look like a convenient way to label data. But there is a deeper reason they exist, and it has to do with **speed**.

Imagine you have a list of 10,000 instrument IDs and you need to check whether a particular ID exists. With a list, Python has to check every single element, one by one, until it finds a match (or reaches the end). This is called a **linear scan** -- it takes time proportional to the size of the list.

A dict uses a technique called **hashing** to jump directly to the right location. It does not need to scan through everything.

Let's see the difference.

```python
# Create a large list and dict of instrument IDs
instrument_list = [f"INST-{i:05d}" for i in range(10000)]
instrument_dict = {f"INST-{i:05d}": f"Sensor Type {i % 5}" for i in range(10000)}

# The ID we're looking for (near the end of the list)
target = "INST-09999"

print(f"List has {len(instrument_list)} items")
print(f"Dict has {len(instrument_dict)} items")
```

```python
%%timeit
# Looking up an item in a LIST (linear scan)
target in instrument_list
```

```python
%%timeit
# Looking up an item in a DICT (hash lookup)
target in instrument_dict
```

The dict lookup is dramatically faster. Depending on your machine, you will likely see something like 100 microseconds for the list versus well under 1 microsecond for the dict. That is a difference of two orders of magnitude.

This is the difference between **O(n)** and **O(1)** in computer science notation:

- **O(n)**: time grows linearly with the size of the data. Double the list, double the search time.
- **O(1)**: time is constant regardless of size. A dict with 10 items is just as fast as one with 10 million.

This is not an academic curiosity. When you are processing millions of rows of telemetry data, the choice between a list lookup and a dict lookup can mean the difference between a pipeline that runs in seconds and one that runs for hours. Dicts exist because fast lookups matter.

## Building a summary

Let's put everything together and build a weekly summary from our data.

```python
# Build a comprehensive weekly summary
all_min = [d["min_temp_c"] for d in week_data]
all_max = [d["max_temp_c"] for d in week_data]
all_wind = [d["wind_speed_ms"] for d in week_data]
all_statuses = [d["status"] for d in week_data]

summary = {
    "period": f"Sol {week_data[0]['sol']} to Sol {week_data[-1]['sol']}",
    "num_sols": len(week_data),
    "avg_min_temp": sum(all_min) / len(all_min),
    "avg_max_temp": sum(all_max) / len(all_max),
    "coldest": min(all_min),
    "warmest": max(all_max),
    "avg_wind": sum(all_wind) / len(all_wind),
    "max_wind": max(all_wind),
    "nominal_count": all_statuses.count("nominal"),
    "warning_count": all_statuses.count("warning"),
    "offline_count": all_statuses.count("offline"),
}

print(f"=== WEEKLY SUMMARY: {summary['period']} ===")
print(f"Sols covered:      {summary['num_sols']}")
print(f"Avg min temp:      {summary['avg_min_temp']:.1f}C")
print(f"Avg max temp:      {summary['avg_max_temp']:.1f}C")
print(f"Coldest reading:   {summary['coldest']}C")
print(f"Warmest reading:   {summary['warmest']}C")
print(f"Avg wind speed:    {summary['avg_wind']:.1f} m/s")
print(f"Max wind speed:    {summary['max_wind']} m/s")
print(f"Instrument status: {summary['nominal_count']} nominal, "
      f"{summary['warning_count']} warning, {summary['offline_count']} offline")
```

## Summary

We have covered a lot of ground:

- **Lists** hold ordered sequences of values. Use them for collections of similar items.
- **Dicts** hold key-value pairs. Use them when data has labels, or when you need fast lookups.
- **for loops** process every item in a collection.
- **List comprehensions** are a compact, Pythonic way to transform and filter lists.
- **Nested structures** (list of dicts) represent tabular data naturally.
- **Dict lookups are O(1)**, list scans are O(n). This matters enormously at scale.

We are starting to process real-ish data now, but we are writing a lot of one-off code. Next, we will learn to wrap our logic in reusable **functions** and read data from **files** -- because nobody wants to type 5,000 sol readings by hand.
